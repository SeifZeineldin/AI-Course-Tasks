<h1>How to combine more than 2 Dataframes</h1>

```python
from functools import reduce
df1 = pd.DataFrame({'team': ['A', 'B', 'C', 'D'],
                    'points': [18, 22, 19, 14]})
df2 = pd.DataFrame({'team': ['A', 'B', 'C'],
                    'assists': [4, 9, 14]})
df3 = pd.DataFrame({'team': ['C', 'D', 'E', 'F'],
                    'rebounds': [10, 17, 11, 10]})
df4 = pd.DataFrame({'team': ['C', 'D', 'E', 'F'],
                    'score': [100, 170, 110, 80]})
dfs = [df1, df2, df3,df4]
final_df = reduce(lambda  left,right: pd.merge(left,right,on=['team'],
                                            how='outer'), dfs)
final_df
```
