```python
import requests
from bs4 import BeautifulSoup
url="https://accounts.google.com/ServiceLogin/signinchooser?service=youtube&uilel=3&passive=true&continue=https%3A%2F%2Fwww.youtube.com%2Fsignin%3Faction_handle_signin%3Dtrue%26app%3Ddesktop%26hl%3Dar%26next%3Dhttps%253A%252F%252Fwww.youtube.com%252Ffeed%252Fchannels%253Fapp%253Ddesktop%26feature%3Dredirect_login&hl=ar&flowName=GlifWebSignIn&flowEntry=ServiceLogin"
header={
    'User_Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.67 Safari/537.36'
}
page=requests.get(url,headers=header).text
html_page=BeautifulSoup(page,"lxml")
print(html_page.prettify())
```
